```python
ASINno = 0
count = 0
for ASIN in ASIN_List:
    count += 1
    if count >= 5:
        driver.close()
        driver = webdriver.Chrome()
        count = 0
    urlPart2 = ASIN
    urlPart4 = 0
    ASINno += 1
    page = 1
    print('\nStarted working on ASIN', ASINno, ':', ASIN)
    
    while 0==0:
        urlFull = urlPart1 + urlPart2 + urlPart3 + str(urlPart4)
        print('\tPage', page, urlFull)
    
        # Load the full page
        driver.get(urlFull)
        time.sleep(1)
        
        html = driver.page_source
        soup = BeautifulSoup(html)
        
        # Find all rows (each row is a seller)
        rowList = soup.find_all('div', class_='a-row a-spacing-mini olpOffer')
        
        for row in rowList:
            # Under each row: name of seller & selling price
            info = row.find('span', class_='a-offscreen').get_text().replace('\n','').replace(',','')
            price = info[info.index("$") : ]
            seller = info[ : info.index("$")].replace('from seller ','').replace(' and price ','')
            
            # Under each row: rating of seller
            try:
                rating = row.find('p', class_='a-spacing-small').get_text().replace('\n','').replace(',','')
                star = rating[ : rating.index(' out of')]
                positive = rating[ rating.index('5 stars')+7 : rating.index('%')+1 ]
                total = rating[ rating.index('(')+1 : rating.index(' total ratings') ]
            except:
                star = '-'
                positive = '-'
                total = '-'

            # Under each row: FBA?
            if row.find('div', class_='olpBadge') != None:
                if 'Fulfillment by Amazon' in row.find('div', class_='olpBadge').get_text():
                    FBA = 'FBA'
                else:
                    FBA = 'Take a look'
            else:
                FBA = '-'

            outputStr = ASIN +','+ seller +','+ price +','+ FBA +','+ star +','+ positive +','+ total +'\n'
            outputFile = open(outputFileName, 'a', encoding='utf-8')
            outputFile.write(outputStr)
            outputFile.close()
        
        # See if we have a NEXT PAGE option
        if soup.find('li', class_='a-last') == None:
            break   
        elif soup.find('li', class_='a-last').find('a') == None:
            break
        else: 
            urlPart4 += 10
            page += 1

print('Job done')
```
